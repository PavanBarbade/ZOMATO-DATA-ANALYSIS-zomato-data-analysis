```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
    
```

```python
dataframe.head()
```

```python
# Assuming 'listed_in(type)' is a column in your dataframe
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result.index, result['votes'],marker = 0, color='green')
plt.xlabel('Types of Restaurant', color='red', size=14)
plt.ylabel('Votes', color='red', size=14)
plt.title('Sum of Votes by Restaurant Type', size=14,)
plt.grid()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'listed_in(type)' is a column in your dataframe
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})

# Plot the data as a column chart
plt.bar(result.index, result['votes'], color='green')

# Add labels and title
plt.xlabel('Types of Restaurant', color='red', size=14)
plt.ylabel('Votes', color='red', size=14)
plt.title('Sum of Votes by Restaurant Type', size=14)

# Add grid lines
plt.grid(axis='y')

# Display the plot
plt.show()
```

```python
plt.hist(dataframe['rate'], bins=10,)
plt.title("rating distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'rate' is a column in your dataframe
# Round the ratings to the nearest integer (optional, depending on your data)
dataframe['rate'] = dataframe['rate'].round()

# Group the data by 'rate' and count the occurrences
grouped_data = dataframe['rate'].value_counts().sort_index()

# Create a DataFrame from the grouped data
result = pd.DataFrame({'count': grouped_data})

# Plot the data as a bar chart
plt.bar(result.index, result['count'], color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Rating', color='black', size=14)
plt.ylabel('Count', color='black', size=14)
plt.title('Rating Distribution', size=14)

# Display the plot
plt.show()
```
